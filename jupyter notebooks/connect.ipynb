## Purpose: Try to find connection between CDC42 and NOTCH1

Specify the input object from Hint module (code for Hint is excluded here)

```python
input_obj = {'entrez': '998',
'name': 'cell division cycle 42',
'symbol': 'CDC42', 'taxonomy': 9606,'umls': 'C1366537',
'display': 'entrez(998) name(cell division cycle 42) symbol(CDC42) taxonomy(9606) umls(C1366537) ',
'type': 'Gene', 'primary': {'identifier': 'entrez', 'cls': 'Gene', 'value': '998'}}
```

Specify the output object from Hint module

```python
output_obj = {'entrez': '4851', 'name': 'notch receptor 1','symbol': 'NOTCH1','taxonomy': 9606,'umls': 'C1334889','display': 'entrez(4851) name(notch receptor 1) symbol(NOTCH1) taxonomy(9606) umls(C1334889) ','type': 'Gene','primary': {'identifier': 'entrez', 'cls': 'Gene', 'value': '4851'}}
```

Import FindConnection module

```python
from biothings_explorer.user_query_dispatcher import FindConnection
```

```python
fc = FindConnection(input_obj, output_obj, registry=reg)
```

Perform query

```python
fc.connect()
```
